# 1 - Qu'est-ce qu'une trame NMEA ?
</div>

Aujourd'hui, les smartphones sont tous dotés d'un récepteur GPS. Ils peuvent se géolocaliser grâce à cette puce, c'est-à-dire  connaître leurs positions géographiques. L'objectif de cette activité est de comprendre le fonctionnement de ce dispositif.

Une trame NMEA est une suite de caractères produite par un récepteur GPS. Cette suite de catactères contient des informations de géolocalisation, comme :
* l'heure de réception
* la latitude
* la longitude
* l'altitude
* le nombre de satellites utilisés pour le calcul
* etc.

Voici un **exemple** de trame NMEA :
`$GPGGA,064036.289,4836.5375,N,00740.9373,E,1,0,43.2200.2,M,,,,,0000*0E`

Décodons cette trame :

| caractères   | signification |
| :---------   | :------------ |
| `$`          | Début d'une nouvelle trame |
| `GP`         | Origine du signal : `GP` pour GPS, `GA` pour Galiléo, `BD` ou `GB` pour Beidu, `GL` pour Glonass, `GN` pour GPS + Glonass |
| `GGA`        | Type de trame : `GGA` pour position (fixe), `RMC` pour navidation (déplacement), ... |
| `064036.289` | Heure de réception : ici à 06h 40m 36,289s (heure UTC) |
| `4836.5375`  | Latitude : ici 48° 36.5375' (DM) = 48.608958° (DD) = 48° 36' 32.25" (DMS) |
| `N`          | `N` pour latitude nord, `S` pour latitude sud |
| `00740.9373` | Longitude : ici 7° 40.9373' (DM) = 7.682288° (DD) = 7° 40' 56.238" (DMS) |
| `E`          | `E` pour longitude est, `W` pour longitude ouest |
| `1`          | Positionnement : `0` pour calé, `1` pour positionné, `2` en mode différentiel, `6` pour estimée |
| `04`         | Nombre de satellites utilisés pour calculer les coordonnées : ici 4 |
| `3.2`        | Précision horizontale : de `1.0` (optimale) à `8.0` non fiable |
| `200.2,M`    | Altitude : ici 200.2 mètres |
| `,,,,,0000`  | D'autres informations peuvent être inscrites dans ces champs |
| `*0E`        | Caractères de contrôle |

Avant tout autre traitement, on pourra commencer par découper la trame au niveau des virgules :
* le 1er morceau contiendra le caractère de départ, l'origine du signal et le type de trame : `$GPGGA`
* le 2e morceau contiendra l'heure de réception : `064036.289`
* le 3e morceau contientra la valeur de latitude : `4836.5375`
* le 4e morceau contientra l'orientation _nord_ ou _sud_ pour la latitude : `N`
* le 5e morceau contiendra la valeur de longitude : `00740.9373`
* le 6e morceau contientra l'orientation _est_ ou _ouest_ pour la longitude : `E`
* etc.

On se limitera ensuite à l'étude des morceaux de chaîne 3 et 4 pour la latitude, et 4 et 6 pour la longitude.

???

!!! info Première étape...

En Python, la méthode `split( )` des chaînes de caractères fait cela parfaitement : 
```python
liste = trame.split(",")
```
Cette commande découpe la chaîne de caractères contenue dans la variable `trame` au niveau des **virgules**. On obtient comme résultat une liste de chaînes, comme autant de morceaux découpés d'un ruban. Les caractères `,` où a eu lieu la coupe ne sont plus présents.

Dans notre exemple, la variable `liste` contiendra ainsi une liste de 15 chaines : `['$GPGGA', '100012.348', '4630.172', 'N', '00147.287', 'W', '1', '12', '1.0', '0.0', 'M', '0.0', 'M', '', '*7B']`
    
Pour obtenir la valeur numérique de la latitude, on récupère le 3e élément de la liste :
```python
latitude = liste[2]   # info : en Python, les indices des éléments commencent à zéro
```
...et pour connaitre son hémisphère, on regarde le 4e élément de la liste :
```python
latitude_hemisph = liste[3]   # info : en Python, les indices des éléments commencent à zéro
```
   
On a ainsi la variable `latitude` qui contient la chaîne de caractères "4630.172", et la variable `latitude_hemisph` qui contient la chaîne de caractères "N".

!!! info Seconde étape
    
Pour le moment, la variable `latitude` contient une chaîne de caractères : "4630.172". On souhaite la transformer en valeur numérique en **degrés décimaux**.
    
Or, dans cette chaîne :
* les 2 premiers caractères contiennent la valeur numérique des degré : `46`
* les caractères 3 jusqu'à la fin contiennent la valeur numérique des minutes d'arc : `30.172`

On extrait ces deux données de manière séparée, puis on les convertit du type _string_ (chaîne de caractères) en _float_ (nombre à virgule) :
```python
latitude_deg = float(latitude[0:2])   # du 0e caractère inclus au 2e caractère exclu, les indices partant de 0
latitude_min = float(latitude[2:])    # du 2e caractère inclus à la fin, les indices partant de 0
```
On peut à présent calculer la valeur de la latitude en degrés décimaux :
```python
lat_dec = latitude_deg + latitude_min / 60
```
Cette fois, la variable `lat_dec` contient bien une valeur numérique de type _float_.<br/>

**C'est maintenant que nous allons étudier si nous sommes dans l'hémisphère Nord ou Sud.**

On doit changer de signe si la latitude est pour l'hémisphère sud : 
```python
if liste[3] == "S":
    lat_dec = -lat_dec
```

On procède de la mème façon pour la longitude. Attention cependant, dans la trame, la valeur des degrés occupe cette fois 3 caractères.
Pour la cardinalité (Est / Ouest), là aussi un changement de signe s'opère si l'on est à l'Ouest du méridien de Greenwhich.

Voici un code Python complet :
```python
trame = "$GPGGA,100012.348,4630.172,N,00147.287,W,1,12,1.0,0.0,M,0.0,M,,*7B"

liste = trame.split(",")

# extraction de la latitude et calcul de sa valeur en degrés décimaux
latitude = liste[2]   # rappel : en Python, les indices des éléments commencent à zéro
latitude_deg = float(latitude[0:2])   # du 0e caractère inclus au 2e caractère exclus, les indices partant 0
latitude_min = float(latitude[2:])    # du 2e caractère inclus à la fin, les indices partant de 0
lat_dec = latitude_deg + latitude_min / 60
if liste[3] == "S":
    lat_dec = -lat_dec

# extraction de la longitude et calcul de sa valeur en degrés décimaux
longitude = liste[4]   # rappel : en Python, les indices des éléments commencent à zéro
longitude_deg = float(longitude[0:3])   # du 0e caractère inclus au 2e caractère exclu, les indices partant 0
longitude_min = float(longitude[3:])    # du 2e caractère inclus à la fin, les indices partant de 0
long_dec = longitude_deg + longitude_min / 60
if liste[5] == "W":
    long_dec = -long_dec

print("La latitude vaut", lat_dec, "°")
print("La longitude vaut", long_dec, "°")
```

````python
trame = "$GPGGA,100012.348,4630.172,N,00147.287,W,1,12,1.0,0.0,M,0.0,M,,*7B"

liste = trame.split(",")

# extraction de la latitude et calcul de sa valeur en degrés décimaux
latitude = liste[2]   # rappel : en Python, les indices des éléments commencent à zéro
latitude_deg = float(latitude[0:2])   # du 0e caractère inclus au 2e caractère exclus, les indices partant 0
latitude_min = float(latitude[2:])    # du 2e caractère inclus à la fin, les indices partant de 0
lat_dec = latitude_deg + latitude_min / 60
if liste[3] == "S":
    lat_dec = -lat_dec

# extraction de la longitude et calcul de sa valeur en degrés décimaux
longitude = liste[4]   # rappel : en Python, les indices des éléments commencent à zéro
longitude_deg = float(longitude[0:3])   # du 0e caractère inclus au 2e caractère exclu, les indices partant 0
longitude_min = float(longitude[3:])    # du 2e caractère inclus à la fin, les indices partant de 0
long_dec = longitude_deg + longitude_min / 60
if liste[5] == "W":
    long_dec = -long_dec

print("La latitude vaut", lat_dec, "°")
print("La longitude vaut", long_dec, "°")
````

Vous pouvez créer un lien vers n'importe quelle carte affichée en utilisant une URL sous la forme:

Le paramètre layers de l’URL correspond à code spécial indiquant les couches de la carte affichées. Ce paramètre est optionnel >et s'il est manquant, le style de carte par défaut Standard sera affiché.  
    
>M = Mapnik - Standard tile layer  
>C = Cyclemap - Cycle Map  
>T = Transport Map - Transport Map  
>O = ÖPNVKarte - ÖPNVKarte  
>H = Humanitarian - Humanitarian  
>Y = CyclOSM - CyclOSM

````python
import folium
"""
On crée l'objet m. On lui assigne le résultat de la fonction folium.Map()
à laquelle on passe des coordonnées géographiques
"""
m=folium.Map(location=[lat_dec, long_dec],zoom_start=16)
m
````

!!! tip Astuce

On peut rajouter une "mini-carte" pour faciliter la navigation. Cela fait partie des nombreuses possibilités offertes par _folium_

!!!

````python
from folium import plugins
m = folium.Map(location=(lat_dec, long_dec), zoom_start=16)

minimap = plugins.MiniMap()
m.add_child(minimap)
m
````

!!! info Les différents fonds dans OpenStreetMap

On peut la aussi utiliser les différents fonds fournis par OpenStreetMap

On appelle ça des tuiles. 

Le site de référence de la librairie Folium  
https://python-visualization.github.io/folium/latest/  
et pour les infos concernant les fonds ('tiles')  
https://python-visualization.github.io/folium/latest/user_guide/raster_layers/tiles.html  

Comment faire l'essai ? 
!!!

````python
folium.Map(
    location=[lat_dec, long_dec],tiles='Cartodb Positron',zoom_start=16
)
````

!!! info Mise en place de marqueurs 
A présent, on souhaiterait personnaliser la carte avec des marqueurs de lieux visités, par exemple... Comment faire ?
